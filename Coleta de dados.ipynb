```python
import youtube_dl
import pandas as pd
import numpy as np
import re
```

```python
queries = ["computer+architecture", "computer+organization", "computer+organization+and+architecture"]
```

```python
ydl = youtube_dl.YoutubeDL({"ignoreerrors": True, "islive": False})
```

```python
results = []
for query in queries:
    raw_data = ydl.extract_info("ytsearchdate2000:{}".format(query), download=False)
    for entry in raw_data['entries']:
        if entry is not None:
            entry['query'] = query
    results += raw_data['entries']
```

```python
results = [e for e in results if e is not None]
len(results)
```

```python
df = pd.DataFrame(results)

df['upload_date'] = pd.to_datetime(df["upload_date"])
df['tempo_desde_pub'] = (pd.to_datetime("2020-07-01") - df['upload_date']) / np.timedelta64(1, 'D') 
```

```python
number_of_columns = df.shape[1]
pd.set_option("display.max_columns", number_of_columns)
df.head()
```

```python
columns = ['title', 'upload_date', 'view_count']

df[columns].to_feather("raw_data.feather")
```

```python
df.to_csv('raw_data_without_labels.csv', index=False)  
```
